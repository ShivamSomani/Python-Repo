```python
import pandas as pd
import numpy as np
from textblob import TextBlob
```

```python
X = pd.read_csv('@realdonaldtrumpfile.csv')
```

```python
X.head()
```

```python
X.tail()
```

```python
tweets = X['Tweet Text']
```

```python
tweets = np.array(tweets)
```

```python
scores = []
for tw in tweets:
    t_scores = TextBlob(str(tw)).sentiment.polarity
    scores.append(t_scores)
```

```python
scores
```

```python
for i in range(len(scores)):
    if scores[i] <=1 and scores[i] > 0.6:
        scores[i] = 2
    elif scores[i] <=0.6 and scores[i] > 0.2:
        scores[i] = 1
    elif scores[i] <= 0.2 and scores[i] > -0.2:
        scores[i] = 0
    elif scores[i] <= -0.2 and scores[i] > -0.6:
        scores[i] = -1
    elif scores[i] <= -0.6 and scores[i] > -1:
        scores[i] = -2
```

```python
len(scores)
```

```python
X['Sentiment Scores'] = scores
```

```python
X.tail()
```

```python
X.to_csv('@realdonaldtrumps_sentiments.csv')
```
