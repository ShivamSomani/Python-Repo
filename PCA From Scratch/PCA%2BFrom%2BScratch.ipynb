```python
import numpy as np
np.random.seed(1) #for debugging, randomly generated numbers remain the same
```

```python
#Creating the dataset
means = np.array([0,0,0]) #sample mean
cov = np.array([[1,0,0],[0,1,0],[0,0,1]]) #sample covariance
#Generating random data using the Gaussian Distribution
sample = np.random.multivariate_normal(means,cov,20).T
print sample.T
#its a 20 x 3 matrix
```

```python
means2 = np.array([1,1,1]) #sample mean
cov2 = np.array([[1,0,0],[0,1,0],[0,0,1]]) #sample covariance
#Generating random data using the Gaussian Distribution
sample2 = np.random.multivariate_normal(means2,cov2,20).T
print sample2.T
#its a 20 x 3 matrix
```

```python
from matplotlib import pyplot as plt
#figure, width and height
fig = plt.figure(figsize=(10,10))
#3d subplot and subplot grid parameters
from mpl_toolkits.mplot3d import Axes3D
ax = fig.add_subplot(111,projection='3d')
#plot samples
ax.plot(sample[0,:],sample[1,:],sample[2,:],
       'o',markersize=8,color='blue',alpha=0.5,label='class1')
ax.plot(sample2[0,:],sample2[1,:],sample2[2,:],
       '^',markersize=8,color='green',alpha=0.5,label='class2')
ax.legend(loc='upper right')
plt.show()
```

```python
#merge the data into one dataset
all_samples = np.concatenate((sample,sample2), axis=1)
print all_samples.T #40 x 3 dataset
```

```python
#Computing the dimensional mean vector which helps in the computation
#of the covariance matrix
#Finding the mean of each feature
mean_x = np.mean(all_samples[0,:])
mean_y = np.mean(all_samples[1,:])
mean_z = np.mean(all_samples[2,:])
#3d mean vector
mean_vector = np.array([[mean_x],[mean_y],[mean_z]])
print mean_vector
```

```python
#Computing the covariance matrix from the mean
#It models the relationship b/w all the different features or the variables
#Variance is the degree by which a random variable changes with respect to its expected value
#CoVariance is the degree by which 2 random variables change with respect to each other
#Covariance describes the relationship between the variables
cov_matrix = np.zeros((3,3))
#Iterating n-1 or 39 times
#all_samples is a 3 x 40 matrix
for i in range((all_samples.shape[1])):
    cov_matrix += (all_samples[:,i].reshape(3,1) - mean_vector).dot((all_samples[:,i].reshape(3,1) - mean_vector).T)
#Actual Matrix is (1/m)*cov_matrix
cov_matrix = cov_matrix/39
print cov_matrix
```

```python
#Compute eigenvectors and eigenvalues
#using the numpy function to find eigen values and eigen vectors
eig_val, eig_vec = np.linalg.eig(cov_matrix)
print 'Values:',eig_val
print 'Vectors:',eig_vec
```

```python
#Create eigenpair from both and sort them
#Firstly make a list of the eigen values and eigen vector tuples
eig_pairs = [(np.abs(eig_val[i]), eig_vec[:,i])
            for i in range(len(eig_val))]
#Sort the (eigenvalue,eigenvector) tuples from high to low
eig_pairs.sort()
eig_pairs.reverse()
#Printing the values of the pairs
print eig_pairs
```

```python
#Choosing the k eigen vectors
#Stacking the arrays in sequence horizontally and choosing 2 features
#Taking the tuples with 2 highest eigen values
final_res = np.hstack((eig_pairs[0][1].reshape(3,1),eig_pairs[1][1].reshape(3,1)))
print final_res
```

```python
#Transforming the data using the eigen pair
#Computing 2 x 3 and 3 x 40
transformed = final_res.T.dot(all_samples)
print transformed.T
```

```python
plt.plot(transformed[0,0:20], transformed[1,0:20],
         'o', markersize=7, color='green', alpha=0.5, label='class1')
plt.plot(transformed[0,20:40], transformed[1,20:40],
         '^', markersize=7, color='red', alpha=0.5, label='class2')
plt.xlim([-5,5])
plt.ylim([-5,5])
plt.xlabel('x_values')
plt.ylabel('y_values')
plt.legend()
plt.title('Transformed samples with class labels')

plt.show()
```
