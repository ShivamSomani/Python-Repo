```python
#Importing the numpy library and the iris dataset
import numpy as np
from sklearn.datasets import load_iris
```

```python
#Importing the matplotlib library for visualization, counter for counting the frequency of the most common 
#observation and the square root for calculating the minimum distance
from math import sqrt
import matplotlib.pyplot as plt
from collections import Counter
```

```python
#Predict function for the KNN Algorithm, parameters are the training data, training labels, the test value for
#which the class needs to be predicted and the number of neighbors to be considered for the prediction of
#the class
def predict(X_train,y_train,testValue,k):
    #Initialization of the lists for distance and targets
    distances = []
    targets = []
    #Calculating the distance of the test value from each of the points in the dataset and storing that distance
    #along with the index of the point in the distance list
    for i in range(len(X_train)):
        dist = np.sqrt(np.sum(np.square(testValue - X_train[i, :])))
        distances.append([dist,i])
    #Sorting the distance list according to the distances
    distances.sort()
    #Initialization of the list for finding the nearest neighbors
    neighbors = []
    #Finding the index of the k-nearest neighbors in the sorted distance list and storing
    #the class label for those neighbors in the neighbors list
    for x in range(k):
        index = distances[k][1]
        neighbors.append(y_train[index])
    #Counting the most frequently occurring class label and returning that value
    vote_result = Counter(neighbors).most_common(1)[0][0]
    return vote_result
#K-Nearest Neighbors function which runs the the predict function for each of the points in the
#testing set and returns the list of predicted class labels
def k_nearest_neighbors(X_train,y_train,X_test,k):
    prediction = []
    for i in range(len(X_test)):
        prediction.append(predict(X_train,y_train,X_test[i,:],k))
    return prediction
```

```python
#Importing the iris dataset and other functions for splitting the data and calculation of accuracy
iris = load_iris()
X = iris.data
y = iris.target
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
```

```python
#Splitting the data with 60% in training and 40 % in test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=2)
scores = []
for i in range(1,15):
    pred = k_nearest_neighbors(X_train,y_train,X_test,i)
    accuracy = accuracy_score(y_test,pred)
    scores.append(accuracy)
xlabels = range(1,15)
plt.plot(xlabels,scores,'ro-')
plt.xlabel('No. of Neighbors')
plt.ylabel('Accuracy Score')
plt.title('Graph for 60-40 split')
plt.show()
```
